## 1. Introdução ao Aprendizado de Máquina

- IA: qualquer técnica que permita as máquinas imitar o comportamento humano
- Aprendizado de Máquina: habilidade de aprender sem ser explicitamente programado para a tarefa
- Aprendizado Profundo: extração de padrões de dados usando redes neurais


```bash
Aprendizado de Máquina
├─ Supervisionado
│   ├─ Classificação
│   └─ Regressão
│
├─ Não-Supervisionado
│   ├─ Agrupamento (*Clustering*)

```

### Aprendizado Supervisionado

- uso de conjunto de dados rotulados para treinar o modelo
- Requerimentos:
  - dados rotulados para treinamento e teste
  - métrica para avaliar o desempenho do modelo
- Tipos de problemas: Classificação e Regressão

### Não-Supervisionado

- descobrir subgrupos (exemplos similares) em um conjunto de dados sem rótulos
- Requerimentos:
  - dados de treinamento para treino e teste
  - NÃO requer variável resposta

## 3. Sobre a Convergência das Redes Neurais

- Modelos de *deep learning* são inspirados na estrutura biológica, mas são construções matemáticas, não são cérebros artificiais

- Neurônio Artificial: abstrai a sinapse através de um peso (força de conexão) e gera um valor de ativação como saída

### Perceptron

- função algébrica comum com entrada e saída
- esses neurônios artificiais são dispostos ao longo de camadas que interagem entre si, criando a estrutura conhecida como Rede Neural ou *Multilayer Perceptron*

- Camada de Entrada
- Camada(s) Intermediária(s) (*Hidden Layer*)
- Camada de Saída

### Arquiteturas

### Neurônio Artificial

#### Entrada
- vetor da vase de dados ou saída dos neurônios da camada anterior ($x_i$)

#### Saída
- combinação não-linear das entradas
  - cada ligação tem um peso associado ($w_i$)
  - cada neurônio tem um viés ($b$)
- a saída do neurônio é dada por $\hat{y} = \sigma(\sum w_i x_i + b) = \sigma(w_1 x_1 + w_2 x_2 + w_3 x_3 + ... + b)$

- um neurônio faz uma transformação afim dos dados de entrada e em seguida aplica uma função de ativação $\sigma$
